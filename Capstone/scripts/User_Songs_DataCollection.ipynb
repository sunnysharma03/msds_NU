```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd
from collections import Counter
```

```python
userName = 'Sachin Sharma'
# Replace these with your Spotify Developer credentials
CLIENT_ID = ''
CLIENT_SECRET = ''
REDIRECT_URI = 'https://accounts.spotify.com/api/token'

# Scopes for accessing user's recently played tracks and top tracks
SCOPE = 'user-read-recently-played'
user_name = 'Sachin Sharma'
```

```python
# Authentication
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=CLIENT_ID,
                                               client_secret=CLIENT_SECRET,
                                               redirect_uri=REDIRECT_URI,
                                               scope=SCOPE))
```

```python
# Fetch user's recently played tracks (last 30 days)
results = sp.current_user_recently_played(limit=50)
recent_tracks = results['items']
```

```python
# Function to get audio features for a track by track ID
def get_audio_features_by_track_id(track_id):
    audio_features = sp.audio_features(track_id)[0]
    return audio_features
```

```python
# Function to get genres for an artist by artist ID
def get_genres_by_artist_id(artist_id):
    artist = sp.artist(artist_id)
    if len(artist['genres']) > 0:
        return artist['genres'][0]
    return ''
```

```python
# Convert track data to a DataFrame
track_data = []
for item in recent_tracks:
    track = item['track']
    artist_id = track['artists'][0]['id']
    genres = get_genres_by_artist_id(artist_id)
    audio_features = get_audio_features_by_track_id(track['id'])
    track_info = {
        'user_name': user_name,  # Replace with the actual username variable if needed
        'artist_name': track['artists'][0]['name'],
        'track_name': track['name'],
        'track_id': track['id'],
        'popularity': track['popularity'],
        'year': track['album']['release_date'][0:4],  # year
        'played_at': item['played_at'],
        'genre': genres,  # list of genres
        'danceability': audio_features['danceability'],
        'energy': audio_features['energy'],
        'key': audio_features['key'],
        'loudness': audio_features['loudness'],
        'mode': audio_features['mode'],
        'speechiness': audio_features['speechiness'],
        'acousticness': audio_features['acousticness'],
        'instrumentalness': audio_features['instrumentalness'],
        'liveness': audio_features['liveness'],
        'valence': audio_features['valence'],
        'tempo': audio_features['tempo'],
        'duration_ms': audio_features['duration_ms'],
        'time_signature': audio_features['time_signature']
    }
    track_data.append(track_info)
recent_tracks_df = pd.DataFrame(track_data)
```
