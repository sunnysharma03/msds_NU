### Preprocessing

few additional refinements to address common preprocessing challenges in text data:

1. Synonym Mapping: Replace synonyms (e.g., "good," "great," "excellent") with a unified term to reduce semantic redundancy.
2. Named Entity Standardization: For words with multiple common forms, standardize terms to prevent fragmentation (e.g., "movie" and "film" could both become "movie").
3. Handle Negations: Combine terms following negation words (e.g., "not good" → "not_good") to retain sentiment nuances.

```python
# Synonym mapping for common terms to improve clustering
synonym_dict = {
    "good": "positive",
    "great": "positive",
    "excellent": "positive",
    "bad": "negative",
    "terrible": "negative",
    "movie": "film",
    "film": "film",
    "picture": "film"
}

def preprocess_text(text):
    # Convert text to lowercase and tokenize
    tokens = word_tokenize(text.lower())
    
    # Remove stopwords and non-alphabetical tokens
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    
    # Lemmatize tokens
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # Apply synonym mapping
    tokens = [synonym_dict.get(word, word) for word in tokens]
    
    # Handle negations by joining "not" with the following word, if present
    tokens = ['_'.join([tokens[i], tokens[i+1]]) if tokens[i] == 'not' and i+1 < len(tokens) else tokens[i] 
              for i in range(len(tokens) - 1) if tokens[i] != 'not'] + ([tokens[-1]] if tokens[-1] != 'not' else [])
    
    return ' '.join(tokens)
```

```python
data['Review'] = data['Review'].apply(preprocess_text)
```

```python
def find_optimal_clusters(tfidf_matrix):
    inertia = []
    for k in range(1, 10):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(tfidf_matrix)
        inertia.append(kmeans.inertia_)

    plt.plot(range(1, 10), inertia, marker='o')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia (WCSS)')
    plt.title('Elbow Plot for Optimal k')
    plt.show()
```

## Clustering with TF-IDF

```python
tf_idf = TfidfVectorizer(max_df = 0.9, min_df = 2, stop_words = 'english')
```

```python
dtm = tf_idf.fit_transform(data['Review'])
```

```python
nmf_model = NMF(n_components=4, random_state=42, max_iter=400)
nmf_model.fit(dtm)
```

```python
for i, topic in enumerate(nmf_model.components_):
    print(f'The top 15 words for topic #{i}')
    print([tf_idf.get_feature_names_out()[index] for index in topic.argsort()[-15:]])
    print("\n\n")
```

```python
k = 2  # Number of clusters
kmeans = KMeans(n_clusters=2, random_state=42)
nmf_features = nmf_model.transform(dtm)
cluster_labels = kmeans.fit_predict(nmf_features)
```

```python
# Evaluate the clustering performance using Silhouette Score
silhouette_avg = silhouette_score(nmf_features, cluster_labels)
print(f'Silhouette Score: {silhouette_avg}')

nmf_ari_score = adjusted_rand_score(data['Sentiment'].apply(lambda x: 1 if x == 'Positive' else 0), cluster_labels)
print(f"Adjusted Rand Index with NMF 4: {nmf_ari_score}")
```

```python
# Add cluster labels to the dataframe
data['nmf_cluster'] = cluster_labels

# Display the number of reviews in each cluster
print(data['nmf_cluster'].value_counts())
```

Given that the clustering is showing a significant imbalance <b>(9 reviews in one cluster and only 1 in the other)</b>, there are a few areas to investigate and refine to improve the clustering so it aligns more closely with the sentiment labels. Here’s an outline of the issues and refinement suggestions:

<b>Observations and Issues</b>
Imbalanced Clusters: With 9 reviews clustered together and only 1 review in a separate cluster, K-means is likely treating most reviews as similar, failing to differentiate between positive and negative sentiments.
NMF Features: NMF (Non-negative Matrix Factorization) can sometimes struggle with fine-grained sentiment distinctions if the input representation does not effectively capture the nuance in positive vs. negative sentiments.

<h3>Steps for Refinement</h3>

1. <b>Increase n_components in NMF:</b> Instead of n_components=4, we can try a higher number, such as n_components=6 or n_components=8, to capture more distinct topics and potentially reveal positive vs. negative sentiment separation.
2. <b>Refine Clustering with Sentiment Labels in Mind:</b> The imbalance might be mitigated by adding a sentiment-based pre-clustering step or tuning parameters to enhance feature separability.
3. <b>Alternate Clustering with DBSCAN:</b> Since K-means struggled to create balanced clusters, DBSCAN (Density-Based Spatial Clustering of Applications with Noise) might work better as it can help identify core and outlier points, potentially splitting positive and negative sentiments more effectively.
Let's implement these steps to try improving the results.

Increasing NMF Components: This should help capture more nuances, potentially separating topics aligned with sentiment.

```python
# Increase the number of NMF components
nmf_model = NMF(n_components=6, random_state=42, max_iter=400)
nmf_model.fit(dtm)
nmf_features = nmf_model.transform(dtm)

# Print new topics to see if they align better with sentiment categories
for i, topic in enumerate(nmf_model.components_):
    print(f'Top words for topic #{i}:')
    print([tf_idf.get_feature_names_out()[index] for index in topic.argsort()[-15:]])
    print("\n")
```

Using the refined NMF features with K-means, let’s check if clusters are more balanced with k=2.

```python
# Apply K-means clustering on new NMF features
kmeans = KMeans(n_clusters=2, random_state=42)
cluster_labels = kmeans.fit_predict(nmf_features)

# Evaluate the clustering performance using Silhouette Score
silhouette_avg = silhouette_score(nmf_features, cluster_labels)
print(f'Silhouette Score after refining NMF: {silhouette_avg}')

# Assign cluster labels to the data and check cluster distribution
data['nmf_cluster_refined'] = cluster_labels
print(data['nmf_cluster_refined'].value_counts())
```

To assess the alignment with sentiment, we’ll compute the Adjusted Rand Index (ARI) and identify any mismatched documents.

```python
# Evaluate clustering vs. sentiment alignment
ari_score = adjusted_rand_score(data['Sentiment'].apply(lambda x: 1 if x == 'Positive' else 0), data['nmf_cluster_refined'])
print(f"Adjusted Rand Index (Cluster vs. Sentiment) after refinement: {ari_score}")

# Print misclustered reviews
mismatched = data[data['nmf_cluster_refined'] != data['Sentiment'].apply(lambda x: 1 if x == 'Positive' else 0)]
print("Mismatched reviews:")
print(mismatched[['Review', 'Sentiment', 'nmf_cluster_refined']])
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=2)
dbscan_labels = dbscan.fit_predict(nmf_features)

# Evaluate the clustering performance using Silhouette Score
silhouette_avg_db = silhouette_score(nmf_features, dbscan_labels)
print(f'Silhouette Score with DBSCAN: {silhouette_avg_db}')

# Evaluate DBSCAN clustering performance with sentiment alignment
dbscan_ari_score = adjusted_rand_score(data['Sentiment'].apply(lambda x: 1 if x == 'Positive' else 0), dbscan_labels)
print(f"Adjusted Rand Index with DBSCAN: {dbscan_ari_score}")

# Assign DBSCAN labels to the data for analysis
data['dbscan_cluster'] = dbscan_labels
print(data['dbscan_cluster'].value_counts())
```

<h3>Analyzing Refinements</h3>
With the increased NMF components from 4 to 6, we hoped to capture more nuanced subtopics that might help differentiate positive and negative sentiments within the reviews. The K-means clustering yielded slightly more balanced clusters (0: 7 and 1: 3), but the Silhouette Score (0.118) and the Adjusted Rand Index (ARI) score (0.071) remain low, indicating that the clustering still does not align well with the true sentiment classes.

Using DBSCAN with eps=0.5 and min_samples=2, we found that it failed to group data effectively, resulting in multiple small clusters (3 distinct clusters) and outliers (-1 label), leading to a negative ARI score (-0.111). DBSCAN typically requires more data points or well-separated clusters, so it is often less effective for short text data like this.

<h3>Cluster Misallignment</h3>
The three mismatched reviews from K-means clustering (Reviews 2, 3, 8) in cluster 0 were labeled as positive but placed within the same cluster as negatives. Here are some key observations from the top words in each topic for these cases:

Topic Words Overlap Across Sentiment Contexts: Words like “dark,” “film,” “story,” and “scene” appear in both positive and negative contexts, making it hard for the clustering algorithm to distinguish sentiment without further context.
Bi-gram Limitations: Although bigrams helped, phrases like “positive batman” or “negative people” may still lack the expressive strength to represent sentiment distinctly in the feature space. This is evident as some topics don’t clearly separate positive and negative opinions.

<h3>Vocabulary Issues and Polysemy</h3>

The terms “dark,” “time,” “scene,” etc., do not consistently signal either positive or negative sentiment due to polysemy (multiple meanings) and context-dependency. For instance:

“Dark” can imply a gritty atmosphere (often positive in film reviews) or imply negativity if associated with “depressing.”
Positive words in some contexts can be neutral in others, e.g., “positive batman” or “good cinematography” are less clear indicators.

## Next level refinements

Update TF-IDF Vectorization with Bi-grams

```python
# TF-IDF with bigrams for improved feature capture
tf_idf = TfidfVectorizer(max_df=0.9, min_df=2, stop_words='english', ngram_range=(1, 2))
dtm = tf_idf.fit_transform(data['Review'])
```

```python
nmf_model_bigrams = NMF(n_components=6, random_state=42, max_iter=400)
nmf_features_bigrams = nmf_model_bigrams.fit_transform(dtm)

# Store cluster labels in the DataFrame
data['nmf_cluster_bigrams'] = nmf_features_bigrams.argmax(axis=1)
```

Apply Doc2Vec for Improved Semantic Representation

```python
# Prepare data for Doc2Vec
tagged_data = [TaggedDocument(words=review.split(), tags=[str(i)]) for i, review in enumerate(data['Review'])]
doc2vec_model = Doc2Vec(tagged_data, vector_size=50, window=2, min_count=1, workers=4, epochs=40)

# Create document vectors
doc_vectors = [doc2vec_model.dv[str(i)] for i in range(len(tagged_data))]
```

Cluster with K-means using Doc2Vec Features

```python
# Use K-means on the Doc2Vec document vectors
kmeans = KMeans(n_clusters=2, random_state=42)
doc2vec_labels = kmeans.fit_predict(doc_vectors)

# Evaluate clustering quality
ari_score = adjusted_rand_score(data['Sentiment'].apply(lambda x: 1 if x == 'Positive' else 0), doc2vec_labels)
silhouette_avg = silhouette_score(doc_vectors, doc2vec_labels)

print(f'Adjusted Rand Index (Doc2Vec): {ari_score}')
print(f'Silhouette Score (Doc2Vec): {silhouette_avg}')
```

```python
# Store the K-means Doc2Vec cluster labels in the data frame
data['doc2vec_cluster'] = doc2vec_labels

# Display the clustering results with the document text and sentiment
print(data[['Review', 'Sentiment', 'doc2vec_cluster']])

# Cluster Distribution
print("\nCluster distribution in doc2vec clustering:")
print(data['doc2vec_cluster'].value_counts())
```

Let's refine the Doc2Vec model by adjusting its parameters. We’ll increase vector_size to capture more nuanced differences in document content and experiment with the window size to see if a broader context helps with sentiment alignment in clusters. Here’s how we can proceed:

Refine Doc2Vec Parameters: Increase vector_size and adjust window size.
Cluster with Refined Vectors: Apply K-means again and assess clustering quality with Adjusted Rand Index (ARI) and Silhouette Score.
Store and Analyze Results: Add refined cluster labels to the data for further evaluation.

```python
# Prepare data for refined Doc2Vec
tagged_data = [TaggedDocument(words=review.split(), tags=[str(i)]) for i, review in enumerate(data['Review'])]

# Train Doc2Vec model with refined parameters
doc2vec_model = Doc2Vec(tagged_data, vector_size=100, window=5, min_count=1, workers=4, epochs=50)

# Create document vectors with refined model
doc_vectors_refined = [doc2vec_model.dv[str(i)] for i in range(len(tagged_data))]

# Use K-means clustering on the refined Doc2Vec document vectors
kmeans_refined = KMeans(n_clusters=2, random_state=42)
doc2vec_labels_refined = kmeans_refined.fit_predict(doc_vectors_refined)

# Evaluate clustering quality
ari_score_refined = adjusted_rand_score(data['Sentiment'].apply(lambda x: 1 if x == 'Positive' else 0), doc2vec_labels_refined)
silhouette_avg_refined = silhouette_score(doc_vectors_refined, doc2vec_labels_refined)

# Store refined cluster results
data['doc2vec_cluster_refined'] = doc2vec_labels_refined

# Display evaluation metrics and cluster distribution
print(f'Adjusted Rand Index (Refined Doc2Vec): {ari_score_refined}')
print(f'Silhouette Score (Refined Doc2Vec): {silhouette_avg_refined}')
print("Cluster distribution in refined Doc2Vec clustering:")
print(data['doc2vec_cluster_refined'].value_counts())
```

```python
# Step 1: Reduce dimensionality to 2D using PCA
pca = PCA(n_components=2)
doc_vectors_pca = pca.fit_transform(doc_vectors_refined)

# Step 2: Create a scatter plot of the clusters
plt.figure(figsize=(10, 6))
plt.scatter(doc_vectors_pca[:, 0], doc_vectors_pca[:, 1], c=data['doc2vec_cluster_refined'], cmap='viridis', alpha=0.5)
plt.title('Doc2Vec Clustering Visualization (Refined)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster Label')
plt.grid()
plt.show()
```

```python
# Step 1: Create a DataFrame to hold the reviews by cluster
clustered_reviews = data.groupby('doc2vec_cluster_refined')['Review'].apply(list).reset_index()

# Step 2: Display the reviews in each cluster
for index, row in clustered_reviews.iterrows():
    print(f"Cluster {row['doc2vec_cluster_refined']} reviews:")
    for review in row['Review']:
        print(f"- {review}")
    print("\n")
```

<h3>Initial Issues with Simple Clustering</h3>
The clustering process did not yield the desired results for several reasons, which fall into two main categories:

1. Term Variants and Entity Issues

    - Inconsistent Treatment of Similar Terms: Terms that should be semantically identical, like "batmen" and "Batman," or even broader sentiment-related variations, such as "loved" and "love," were treated as distinct by the algorithm. These variations can lead to fragmented clusters, especially in sentiment analysis, where consistent terminology is crucial for distinguishing positive and negative language.

    - Missed Entity Connections: Terms like “Bruce Wayne,” “Batman,” “Gotham,” and “Riddler” hold specific connotations but are sometimes separated in the document-term matrix, resulting in weaker feature connections. These separations affect the model’s ability to identify meaningful themes within reviews, contributing to mixed or incoherent clusters.

2. Concept Representation Limitations

    - Insufficient Conceptual Coverage: Words or phrases that are central to understanding sentiment were not adequately captured in the initial TF-IDF matrix. Phrases like "not recommend" or "strongly disagree" are key indicators of negative sentiment but are treated as separate words, reducing their significance in clustering.
    - Underrepresentation of Contextual Nuance: The TF-IDF model and basic NMF clustering do not capture context or semantic relationships between words. For instance, terms like "dark" and "noir" could indicate a positive review for certain viewers (who enjoy those styles) but could signify negativity for others, leading to sentiment ambiguity in clusters.

<h3>Refinements and Observations</h3>
To address these issues, we implemented several refinements to improve the clustering results:

1. Increased NMF Components to Capture Nuance

    - Adjustment: We increased the NMF components from 4 to 6 to capture a broader range of topics, which we hoped would help separate clusters based on sentiment.
    - Outcome: The adjustment revealed distinct topics but did not achieve clear alignment with sentiment, as shown by the Adjusted Rand Index (ARI), which remained low. This result highlights that while additional topics may capture more thematic nuances, they do not necessarily improve sentiment separation due to overlapping language and concepts.

2. Incorporating Bigrams with TF-IDF

    - Adjustment: To address the issue of phrase continuity (e.g., “not recommend”), we used bigrams in the TF-IDF matrix, allowing the model to capture more meaningful sentiment expressions.
    - Outcome: The bigram-enhanced TF-IDF provided some improvement in feature representation, but the clusters still showed mixed sentiment alignment. This approach helped identify recurring phrases, yet without context, these phrases were insufficient for reliable sentiment-based clustering.

3. Doc2Vec with Refined Parameters

    - Adjustment: We introduced Doc2Vec with refined parameters (increased vector_size and window) to capture more semantic relationships and account for word contexts, aiming for a more sentiment-oriented clustering.
    - Outcome: Doc2Vec improved the internal coherence of clusters (as shown by the higher Silhouette Score), indicating that documents within clusters were more thematically similar. However, sentiment alignment remained weak, with the ARI score staying low, suggesting that Doc2Vec effectively grouped documents by general tone and topic but failed to consistently map clusters to sentiment.

4. Experimenting with DBSCAN

    - Adjustment: We applied the DBSCAN algorithm, which is density-based and could potentially identify outliers or smaller clusters. DBSCAN’s flexibility in defining clusters by density made it a useful test for isolating unique reviews.
    - Outcome: DBSCAN did identify outliers but produced highly fragmented clusters without strong sentiment alignment, as indicated by a negative ARI score. The result underscores that density-based clustering is insufficient for sentiment clustering when reviews contain overlapping vocabulary but differ primarily in sentiment.
