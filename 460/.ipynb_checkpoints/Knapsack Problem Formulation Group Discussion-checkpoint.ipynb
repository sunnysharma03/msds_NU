```python
from pulp import LpVariable, LpProblem, LpMaximize

# Define variables
bedroom_set = LpVariable("bedroom_set", lowBound=0, upBound=1, cat='Integer')
dining_room_set = LpVariable("dining_room_set", lowBound=0, upBound=1, cat='Integer')
computer = LpVariable("computer", lowBound=0, upBound=1, cat='Integer')
sofa = LpVariable("sofa", lowBound=0, upBound=1, cat='Integer')
tv = LpVariable("tv", lowBound=0, upBound=1, cat='Integer')

# Create the problem
prob = LpProblem("Transport_Optimization", LpMaximize)

# Define objective function
prob += 60 * bedroom_set + 48 * dining_room_set + 14 * computer + 31 * sofa + 10 * tv

# Define constraints
prob += 800 * bedroom_set + 600 * dining_room_set + 300 * computer + 400 * sofa + 200 * tv <= 1100

# Solve the problem
prob.solve()

# Print the solution status
print("Status:", prob.status)

# Print the optimized objective function value
print("Optimized Value: $", value(prob.objective))

# Print decision variables and their values
for var in prob.variables():
    print(f"{var.name} = {var.varValue}")
```
