##### ii) What is the most that Seat and Greet should be willing to pay for an extra cubic foot of stuffing?

```python
# Create the LP problem instance
prob = LpProblem(name="Maximize_Profit", sense=LpMaximize)

# Define decision variables
x1 = LpVariable(name="x1", lowBound=0, cat='Integer')  # Number of couches
x2 = LpVariable(name="x2", lowBound=0, cat='Integer')  # Number of love seats

# Define constraints
prob += 120*x1 + 70*x2 <= 9010  # Fabric constraint
prob += 40*x1 + 25*x2 <= 3500   # Stuffing constraint
prob += 3*x1 + 2*x2 <= 250      # Workers constraint
prob += x1 >= 50                # Minimum couch production
prob += x2 >= 40                # Minimum love seat production

# Define objective function
prob += 850*x1 + 650*x2         # Maximize profit

# Solve the LP problem
prob.solve()

# Print the optimal solution
print(f"Optimal Solution:")
print(f"Number of Couches (x1): {int(x1.value())}")
print(f"Number of Love Seats (x2): {int(x2.value())}")
print(f"Objective Function Value: ${prob.objective.value()}")

# Print the sensitivity analysis output
print("\nSensitivity Analysis Output:")
o = [{'Constraint Name': name, 'Shadow Price': c.pi, 'Slack': c.slack} for name, c in prob.constraints.items()]
print(pd.DataFrame(o))

# Calculate the most Seat and Greet should be willing to pay for an extra cubic foot of stuffing
current_profit_per_stuffing = 650  # Current profit per cubic foot of stuffing for love seats
shadow_price_stuffing = [c.pi for name, c in prob.constraints.items() if name == "_C2"][0]

most_to_pay = shadow_price_stuffing * current_profit_per_stuffing
print(f"\nThe most Seat and Greet should be willing to pay for an extra cubic foot of stuffing: ${most_to_pay:.2f}")
```

##### iv) Seat and Greet is considering producing reclining chairs.  A reclining chair contributes $500 to profit and requires 30 square feet of fabric, 15 cubic feet of stuffing, and 2 workers to produce a chair in one day. Should Seat and Greet produce any reclining chairs? Again, do not rerun the model.

```python
# Shadow Prices from sensitivity analysis
shadow_prices = {
    'Fabric': -0.0,
    'Stuffing': -0.0,
    'Workers': -0.0
}

# Profit per reclining chair and resource requirements
profit_per_reclining_chair = 500
fabric_requirement_per_chair = 30
stuffing_requirement_per_chair = 15
workers_requirement_per_chair = 2

# Calculate change in profit for producing one more reclining chair
change_in_profit = (
    profit_per_reclining_chair +
    shadow_prices['Fabric'] * fabric_requirement_per_chair +
    shadow_prices['Stuffing'] * stuffing_requirement_per_chair +
    shadow_prices['Workers'] * workers_requirement_per_chair
)

# Determine whether Seat and Greet should produce reclining chairs
if change_in_profit > 0:
    print("Seat and Greet should produce reclining chairs.")
    print(f"Change in Profit: ${change_in_profit:.2f}")
else:
    print("Seat and Greet should not produce reclining chairs.")
    print(f"Change in Profit: ${change_in_profit:.2f}")
```

#### Suppose you are in the market to buy a new car for 20,000. The total maintenance costs for this car are dependent on its age in years (see table below). So the total maintenance of a car after 3 years is 7,000, not (3,000 + 5,000 + 7,000 = 15,000). However, you can avoid growing maintenance costs by trading in the car at any point, the value of which is also dependent on its age in years (see same table below). Suppose that if you trade a car in at any point in the next 5 years, the cost of the new car you purchase is still 20,000. You hope to minimize the net cost of having a car over the next five years (purchase costs + maintenance costs – trade-in value).
 
#### Age of car | Maintenance | Trade-In
#### -----------------------------------
#### 0          | 3000       | N/A
#### 1          | 5000       | 12000
#### 2          | 7000       | 10000
#### 3          | 13000      | 6000
#### 4          | 20000      | 2000
#### 5          | N/A        | 1000


#### Part A: Formulate this as a shortest-path network problem and draw the network. As a hint, think about the cost associated with having the car from year 1 to year 2 - $20,000 + $3,000 - $12,000 = $11,000. What is it from year 1 to year 3, or year 2 to year 3?

```python
# Create a directed graph
G = nx.DiGraph()

# Add nodes
nodes = [1, 2, 3, 4, 5, 6]
G.add_nodes_from(nodes)

# Add edges with weights all weights are in 1000's
edges = [
    (1, 2, 11),
    (1, 3, 15),
    (1, 4, 21),
    (1, 5, 31),
    (1, 6, 39),
    (2, 3, 11),
    (2, 4, 15),
    (2, 5, 21),
    (2, 6, 31),
    (3, 4, 11),
    (3, 5, 15),
    (3, 6, 21),
    (4, 5, 11),
    (4, 6, 15),
    (5, 6, 11)
]
G.add_weighted_edges_from(edges)

# Use Dijkstra's algorithm to find the shortest path
shortest_path = nx.dijkstra_path(G, 1, 6)
shortest_path_cost = nx.dijkstra_path_length(G, 1, 6)

# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=1, k=3)  # positions for all nodes

# Nodes
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=700, alpha=0.9)

# Edges
nx.draw_networkx_edges(G, pos, width=2, alpha=0.5, edge_color='black')

# Labels
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Node labels
labels = {node: f'{node}' for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels, font_size=12)

plt.title('Shortest-Path Network Problem')
plt.axis('off')
plt.show()

print(f"The shortest path from node 1 to node 6 is: {shortest_path}")
print(f"The total cost for the shortest path is: ${shortest_path_cost*1000}")
```

#### Question 4
#### An average of 900 cars enter a traffic network (shown below) each hour, trying to make it from node 1 to node 9. The maximum number of cars that can pass over each arc and the time it takes a car to drive each arc is shown below. Use a minimum cost network flow model to minimize the total time for the 900 cars to drive from node 1 to node 9 (assuming all enter at the same time and traffic jams do not exist).	2
8	9	700	7


##### Part A:  Formulate the problem as a minimum cost network flow problem

##### Minimize Z = 16f_12 + 30f_13 + 7f_14 + 17f_25 + 12f_26 + 1f_35 + 4f_36 + 18f_37 + 20f_47 + 13f_56 + 5f_59 + 5f_69 + 2f_78 + 7f_89

#### Part B:  Solve the problem and provide the solution (decision variables and objective function).

```python
prob = LpProblem("Minimum_Cost_Network_Flow", LpMinimize)

arcs = {
    ('1', '2'): {'capacity': 500, 'cost': 16},
    ('1', '3'): {'capacity': 500, 'cost': 30},
    ('1', '4'): {'capacity': 400, 'cost': 7},
    ('2', '5'): {'capacity': 300, 'cost': 17},
    ('2', '6'): {'capacity': 350, 'cost': 12},
    ('3', '5'): {'capacity': 400, 'cost': 1},
    ('3', '6'): {'capacity': 300, 'cost': 4},
    ('3', '7'): {'capacity': 250, 'cost': 18},
    ('4', '7'): {'capacity': 400, 'cost': 20},
    ('5', '6'): {'capacity': 200, 'cost': 13},
    ('5', '9'): {'capacity': 600, 'cost': 5},
    ('6', '9'): {'capacity': 900, 'cost': 5},
    ('7', '8'): {'capacity': 700, 'cost': 2},
    ('8', '9'): {'capacity': 700, 'cost': 7}
}

flow_vars = LpVariable.dicts("Flow", arcs.keys(), lowBound=0, cat=LpContinuous)
for arc, attr in arcs.items():
    flow_vars[arc].upBound = attr['capacity']

#objective function
prob += lpSum([flow_vars[arc] * attr['cost'] for arc, attr in arcs.items()]), "Total Cost of Transport"

#flow conservative constaraints

nodes = set([n for arc in arcs.keys() for n in arc])
for node in nodes:
    if node != '1' and node != '9':
        inflow = lpSum([flow_vars[i, node] for i, j in arcs.keys() if j == node])
        outflow = lpSum([flow_vars[node, j] for i, j in arcs.keys() if i == node])
        prob += (inflow == outflow), f"Flow Conservation at Node {node}"


#supply and demand constarint
prob += lpSum([flow_vars[('1', j)] for i, j in arcs.keys() if i == '1']) == 900, "Supply at Node 1"
prob += lpSum([flow_vars[(i, '9')] for i, j in arcs.keys() if j == '9']) == 900, "Demand at Node 9"

prob.solve()

#output result
print(f"Status: {LpStatus[prob.status]}")

for v in prob.variables():
    print(f"{v.name} = {v.varValue}")

#total cost
print(f"Total travel time: {value(prob.objective)} minutes")
```

```python
'''
A university has three professors who each teach four courses per year.  
Each year, four sections of marketing, finance, and production must be offered.  
At least one section of each class must be offered during each semester (fall and spring). 
Each professor’s time preferences and preference for teaching various courses are given below.
The total satisfaction a professor earns teaching a class is the sum of the semester satisfaction and the course satisfaction.  
Thus, professor 1 derives a satisfaction of 3 + 3 = 6 from teaching marketing during the fall semester


			   Professor 1	Professor 2	  Professor 3
Fall Term		    3		   5		     4
Spring Term 		4		   3             5

Marketing	    	3		   5		     5
Finance		    	7		   4		     7
Production	    	5		   7		     6

'''
```

##### Part A:  Formulate the problem as a minimum cost network flow problem that can be used to assign professors to courses to maximize the total satisfaction of the three professors.  Draw the network and identify the nodes and arcs.

```python
df = pd.DataFrame(np.array([[3,4,3,7,5], [5,3,5,4,7], [4,5,5,7,6]]),
                   index=['Prof1', 'Prof2', 'Prof3'],
                   columns=['Fall Term','Spring Term','Marketing','Finance','Production']).T

inds = df.index.values
cols = df.columns.values

G = nx.Graph()
G.add_nodes_from(df.columns.values)
G.add_nodes_from(df.index.values)

plt.figure(figsize=(10, 10))

# Loop through professors and courses to add edges with lengths based on total satisfaction
for ind in inds:
    for col in cols:
        if ind in ['Fall Term', 'Spring Term']:  # Check if it's a term preference
            total_satisfaction = df.loc[ind, col]
        else:
            total_satisfaction = df.loc[ind, col] + df.loc['Fall Term', col] + df.loc['Spring Term', col]
        G.add_edge(ind, col, length=round(total_satisfaction, 2)) 

pos = nx.spring_layout(G)

nx.draw(G, pos=pos, nodelist=inds[:-2], with_labels=True, node_color='b', node_size=80, font_size=10)
nx.draw(G, pos=pos, nodelist=inds[-2:], with_labels=True, node_color='g', node_size=80, font_size=10)
nx.draw(G, pos=pos, nodelist=cols, with_labels=True, node_color='y', node_size=80, font_size=10)

# Get edge labels
edge_labels = nx.get_edge_attributes(G, 'length')

# Draw edge labels
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.savefig('scheduling.png') 
plt.show()
```
